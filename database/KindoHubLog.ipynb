# [dbo].[fn_GetLogStats]

```sql
/****** Object:  UserDefinedFunction [dbo].[fn_GetLogStats]    Script Date: 23/02/2026 1:36:45 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE FUNCTION [dbo].[fn_GetLogStats]
(
    @HoursBack INT = 24
)
RETURNS TABLE
AS
RETURN
(
    SELECT 
        [Level],
        COUNT(*) as LogCount,
        COUNT(DISTINCT [Username]) as UniqueUsers,
        COUNT(DISTINCT [IpAddress]) as UniqueIPs,
        COUNT(CASE WHEN [Exception] IS NOT NULL THEN 1 END) as ErrorCount,
        MIN([TimeStamp]) as OldestLog,
        MAX([TimeStamp]) as NewestLog
    FROM [dbo].[Logs]
    WHERE [TimeStamp] >= DATEADD(HOUR, -@HoursBack, GETUTCDATE())
    GROUP BY [Level]
);
GO
```

# [dbo].[vw_LogsSummary]

```sql
/****** Object:  View [dbo].[vw_LogsSummary]    Script Date: 23/02/2026 1:36:45 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER OFF
GO

CREATE VIEW [dbo].[vw_LogsSummary]
AS
SELECT 
    [Level],
    COUNT(*) as TotalLogs,
    COUNT(DISTINCT [UserId]) as UniqueUsers,
    COUNT(DISTINCT [IpAddress]) as UniqueIPs,
    MIN([TimeStamp]) as FirstLog,
    MAX([TimeStamp]) as LastLog,
    COUNT(CASE WHEN [Exception] IS NOT NULL THEN 1 END) as LogsWithExceptions
FROM [dbo].[Logs]
GROUP BY [Level];
GO
```

# [dbo].[vw_RecentLogs]

```sql
/****** Object:  View [dbo].[vw_RecentLogs]    Script Date: 23/02/2026 1:36:45 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER OFF
GO

CREATE VIEW [dbo].[vw_RecentLogs]
AS
SELECT TOP 1000
    [Id],
    [TimeStamp],
    [Level],
    [Message],
    [Username],
    [IpAddress],
    [RequestPath],
    [SourceContext],
    [Exception]
FROM [dbo].[Logs]
WHERE [TimeStamp] >= DATEADD(HOUR, -24, GETUTCDATE())
ORDER BY [TimeStamp] DESC;
GO
```

# [dbo].[vw_LoginActivity]

```sql
/****** Object:  View [dbo].[vw_LoginActivity]    Script Date: 23/02/2026 1:36:45 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER OFF
GO

CREATE VIEW [dbo].[vw_LoginActivity]
AS
SELECT 
    [TimeStamp],
    [Username],
    [IpAddress],
    [Message],
    CASE 
        WHEN [Message] LIKE '%Successful login%' THEN 'Success'
        WHEN [Message] LIKE '%Failed login%' THEN 'Failed'
        WHEN [Message] LIKE '%locked out%' THEN 'Locked'
        WHEN [Message] LIKE '%inactive%' THEN 'Inactive Account'
        ELSE 'Other'
    END as LoginStatus,
    [SourceContext]
FROM [dbo].[Logs]
WHERE [Message] LIKE '%login%'
    AND [TimeStamp] >= DATEADD(DAY, -7, GETUTCDATE());
GO
```
